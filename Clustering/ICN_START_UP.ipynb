```python
df2=df.drop('State',axis=1)
df2=df.drop('State',axis=1)
df2
```

```python
df2
```

```python
df2.drop('District', inplace=True, axis=1)
df2
```

```python
arr = df2.to_numpy()
arr
```

```python
len(arr)
```

```python
plt.scatter(arr[:,0], arr[:,1])
```

```python
Kmean = KMeans(n_clusters=33)
Kmean.fit(arr)
```

```python
z=Kmean.cluster_centers_
z
```

```python
plt.scatter(z[:,0], z[:,1])
```

```python
len(z)
```

```python
BBox = ((df.Longitude.min(),   df.Longitude.max(),      
         df.Latitude.min(), df.Latitude.max()))
```

```python
BBox
```

```python
mymap = plt.imread('map.png')
```

```python
ig, ax = plt.subplots(figsize = (8,7))
ax.scatter(df.Longitude, df.Latitude, zorder=1, alpha= 0.2, c='b', s=10)
ax.set_title('Plotting Spatial Data on India Map')
ax.set_xlim(BBox[0],BBox[1])
ax.set_ylim(BBox[2],BBox[3])
ax.imshow(mymap, zorder=0, extent = BBox, aspect= 'equal')
```
